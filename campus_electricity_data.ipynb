# Combine the Datasets

```
frames = [fy17, fy18, fy19]
time_series = pd.concat(frames)
# time_series.time = time_series.index
time_series
```

```
# group by year
grouped = time_series.groupby(pd.to_datetime(time_series['time']).dt.year)
```

```
df17 = grouped.get_group(2017)
df18 = grouped.get_group(2018)
df18 = df18.reset_index(drop=True)
df17 = df17.reset_index(drop=True)

df18.time = df18.index
df17.time = df17.index

df18 = df18[['time','kw']]
df17 = df17[['time','kw']]

df17.to_csv('campus_demand2017.csv')
df18.to_csv('campus_demand2018.csv')
```

```
df17
```

```
fig = plt.figure(figsize=(16,9),edgecolor='k',facecolor='w')
plt.plot(df18['time'],df18['kw'])
plt.title('Electricity Demand for UIUC 2018',fontsize=16)
plt.xlabel('Time [hours]',fontsize=16)
plt.ylabel('Power [kilowatts]',fontsize=16)
plt.show()
```
